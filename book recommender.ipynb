```python
### recommender based on publisher


@interact
def recommend_books_publishers(publisher_name = list(books['publisher'].value_counts().index)):
    a = books[books['publisher'] == publisher_name][['title','average_rating']]
    a = a.sort_values(by = 'average_rating',ascending = False)
    return a.head(20)

```

```python
### recommender based on authors

@interact
def recommend_books_authors(authors_name = list(books['authors'].value_counts().index)):
    a = books[books['authors'] == authors_name][['title','average_rating']]
    a = a.sort_values(by = 'average_rating',ascending = False)
    return a.head(20)


```

```python
### recommender based on language

@interact
def recommend_books_language(language = list(books['language_code'].value_counts().index)):
    a = books[books['language_code'] == language][['title','average_rating']]
    a = a.sort_values(by = 'average_rating',ascending = False)
    return a.head(20)
```

```python
def number_to_obj(x):
    if x >0 and x <=1:
        return 'between 0 and 1'
    if x >1 and x <=2:
        return 'between 1 and 2'
    if x >2 and x <=3:
        return 'between 2 and 3'
    if x >3 and x <=4:
        return 'between 3 and 4'
    if x >4 and x <=5:
        return 'between 4 and 5'
books['rating'] = books['average_rating'].apply(number_to_obj)  
```

```python
books['rating'].value_counts()
```

```python
####encode data
rat=pd.get_dummies(books['rating'])
lang=pd.get_dummies(books['language_code'])
```

```python
rat.head(5)
```

```python
lang.head(5)
```

```python
df = pd.concat([rat,lang, books['average_rating'],books['ratings_count'],books['title']], axis = 1)
df.set_index('title', inplace = True)
```

```python
@interact
def Bookrecommender(book_name =list(books['title'].value_counts().index)):
    book_list_name = []
    book_id = books[books['title'] == book_name].index
    book_id = book_id[0]
    for newid in idlist[book_id]:
        book_list_name.append(books.loc[newid].title)
    return book_list_name    




```
